```python
from kafka import KafkaProducer
from pyspark.sql import SparkSession, DataFrame, Row
from pyspark.sql.functions import to_json
import time
import json
import random
```

```python
path = 'E:\\Downloads\\diabetes_dataset\\online.csv'
```

```python
spark = SparkSession.builder.master("local[*]") \
        .appName('producer').getOrCreate()
```

```python
df = spark.read.csv(path, inferSchema=True, encoding='utf-8', header=True).drop(*['Diabetes_012'])
df.cache()
df.toPandas()
```

```python
producer = KafkaProducer(bootstrap_servers='localhost:9092', security_protocol='PLAINTEXT')
```

```python
for row in df.collect():
    print(json.dumps(row.asDict()))
    producer.send(
        topic="health_data",
        value=json.dumps(row.asDict()).encode('utf-8'))
    time.sleep(random.randint(500, 2000)/ 1000.0)
```
